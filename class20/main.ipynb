```python
def find_unique(arr):
    # Determine the common number using the first three elements
    if arr[0] == arr[1]:
        common = arr[0]
    elif arr[0] == arr[2]:
        common = arr[0]
    else:
        common = arr[1]
    
    # Find the different number
    for i in range(len(arr)):
        if arr[i] != common:
            return arr[i]

# Examples
print(find_unique([1, 1, 1,1,1,1,2]))        # Output: 2
print(find_unique([17, 17, 3, 17, 17, 17, 17]))  # Output: 3
print(find_unique(arr))  # Output: 3
```
