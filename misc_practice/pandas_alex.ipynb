# Filtering and ordering in pandas

### Filltering

```python
df[df['driver_age'] <= 20]
```

#### looking for a specific value or filtering by a specific value 

```python
specific_voilation = ['Registration/plates', 'Equipment']

df[df['violation'].isin(specific_voilation)]
```

#### using contains method

```python
df[df['violation'].str.contains("Equipment")] # having problem but we can check
```

### we can set index 

```python
df2 = df.set_index('stop_time')
df2
```

```python
df2.filter(items=["driver_age","driver_race"]) # default axis= 1 but we can change 
```

```python
df2.filter(items=["driver_age"], axis = 0) # haing some indexing issues but its gonna work on df
```
