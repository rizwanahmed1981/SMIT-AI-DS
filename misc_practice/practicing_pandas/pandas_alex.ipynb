# Order

1) df[df['driver_age'] < 20].sort_values(by='driver_age')
2) by this way we can order the dataframe by one column and accending order

3) df[df['driver_age'] < 20].sort_values(by='driver_age', ascending=False)
4) by making ascending False we can reverse the order in descending order by default it ascending=True

```python

df[df['driver_age'] < 20].sort_values(by='driver_age')
```

```python
df[df['driver_age'] < 20].sort_values(by='driver_age', ascending=False)
```

```python
df[df['driver_age'] < 20].sort_values(by=['driver_age', 'violation'], ascending=False)


## in this case we will have two colums by which data frame is going to be ordered 
# there is one thing noteable that when we pass a list in by=[] it gives prefernce to the first item in the list and sorts the data frame by it then the second item
# we can make it customized like we can pass which one to be sorted and which one not in passing a list of boolean values in ascending=[]
```
