### There is an aggregation function to run multiple aggregations at once

Note: `Base Flavor` is a non-numeric (string) column, so `mean`/`sum` don't make sense for it. Use `agg` on numeric columns or for counts.

Examples:
- df.groupby("Base Flavor").size()  # count rows per group
- df.groupby("Base Flavor").agg({"Texture Rating": ["mean", "max", "min", "sum"]})  # aggregate a specific numeric column

Or aggregate all numeric columns:
- numeric_cols = df.select_dtypes(include="number").columns

- df.groupby("Base Flavor")[numeric_cols].agg(["mean", "max", "min", "sum"])
