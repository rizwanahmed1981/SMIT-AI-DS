## Lets drop all the mising values

#### axis=0. means horizontal drop
#### axis=1, means vertical drop

```python
data.dropna(axis=0, inplace=True)
```

```python
data.isna().sum()
```

```python
data.dropna(axis=1)
```

#### drop only those records where all the values are missing dropna(how='all', axis=0) drops horizontal data is all values are null
#### drop only those records where all the values are missing dropna(how='all', axis=1) drops vertical data is all values are null

```python
# drop only those records where all the values are missing
data.dropna(how='all',axis=0)
```

```python
data.isna().sum()
```

```python
data.shape[1] * 0.70
```

```python
int(data.shape[1] * 0.70)
```

```python
# retain all the records where at least 70% value are valid
data.dropna(axis=0, thresh=int(data.shape[1]*.70)).isnull().sum()
```

```python
# drop all those records where any value missing in year price or kms_driven
data.dropna(axis=0, subset=['year','Price','kms_driven']).isna().sum()
```
