### **Q1. Google â€” Find the Second Largest Number**
Write a Python function to find the second largest number in a list **without using `sort()`**.

```python
def second_largest(nums):

    if (len(nums) < 2):
        return None  # not enough elements

    first_largest = float('-inf')
    second_largest = float('-inf')

    for n in nums:

        if (n > first_largest):

            second_largest = first_largest
            first_largest = n

        elif (first_largest > n > second_largest):
            second_largest = n

    return second_largest if second_largest != float('-inf') else None

# Example
print(second_largest([10, 12, 45, 67, 98]))  # Output: 45
```

### **Q2. Microsoft â€” Count Frequency of Words**
Count how many times each word appears in a given sentence.

```python
def word_count(sentence):
    words = sentence.split()
    freq = {}
    for word in words:
        freq[word] = freq.get(word, 0) + 1
    return freq

print(word_count("Python is easy and Python is powerful"))
```

### **Q3. Amazon â€” Check Palindrome**
Check if a given string is a palindrome.

```python
def is_palindrome(s):
    s = s.lower()
    return s == s[::-1]

print(is_palindrome("Madm"))  # Output: True
```

### **Q4. Google â€” Find Missing Number in List**
Find the missing number from 1 to n in a list.

```python
def find_missing(nums):

    n = len(nums) + 1
    sum1 = n * (n + 1) // 2
    sum2 = sum(nums)
    missing_number = sum1 - sum2
    return missing_number

print(find_missing([1, 2, 4, 6, 3, 7, 8]))  # Output: 5
```

### **Q5. Meta â€” Remove Duplicates from a List**
Remove duplicates without using `set()`.

```python
def remove_duplicates(lst):
    result = []
    for i in lst:
        if (i not in result):
            result.append(i)
    return result

print(remove_duplicates([1, 2, 2, 3, 4, 4, 5]))
```

### **Q6. Microsoft â€” Check if Two Strings Are Anagrams**
Check if two strings are anagrams.

```python
def is_anagram(s1, s2):

    s1 = s1.replace(" ", "").lower()
    s2 = s2.replace(" ", "").lower()

    return sorted(s1) == sorted(s2)

print(is_anagram("Listen", "Silent"))     # True
print(is_anagram("Dormitory", "Dirt room"))  # True
```

### **Q7. Amazon â€” Find Common Elements Between Two Lists**
Return common elements between two lists.

```python
def common_elements(list1, list2):

    common = []

    for i in list1:
      if (i in list2):
        common.append(i)

    return common

print(common_elements([1,2,3,4], [3,4,5,6]))
```

### **Q8. Google â€” Count Vowels in String**
Count how many vowels appear in a string.

```python
def count_vowels(s):

    vowels = "aeiouAEIOU"
    count = 0

    for char in s:
        if (char in vowels):
            count += 1

    return count

print(count_vowels("Google Interview"))
```

### **Q9. Netflix â€” Sum of Digits Using While Loop**
Find the sum of digits of an integer (without converting to string).

```python
def sum_of_digits(n):
    total = 0
    while n > 0:
        total += n % 10
        n //= 10
    return total

print(sum_of_digits(12345))  # Output: 15
```

### **Q10. Google â€” Fibonacci Series up to n Terms**
Print the Fibonacci series.

```python
def fibonacci(n):
    a, b = 0, 1
    for i in range(n):
        print(a, end=" ")
        a, b = b, a + b

fibonacci(7)
```

### **Q11. Microsoft â€” Find Max and Min Without Using Built-ins**
Find the maximum and minimum numbers in a list **without using `max()` or `min()`**.

```python
def find_min_max(lst):

    min_val = lst[0]
    max_val = lst[0]

    for num in lst:
        if (num > max_val):
            max_val = num
        elif (num < min_val):
            min_val = num

    print("List is:", lst)
    print("Max =", max_val)
    print("Min =", min_val)

find_min_max([4, 9, 1, 3, 7])
```

### **Q12. Amazon â€” Count Even and Odd Numbers**
Count how many even and odd numbers are in a list.

```python
def count_even_odd(nums):
    even = odd = 0
    for n in nums:
        if n % 2 == 0:
            even += 1
        else:
            odd += 1

    print("List is:", nums)
    print("Even:", even)
    print("Odd:", odd)

count_even_odd([1, 2, 3, 4, 5, 6])
```
