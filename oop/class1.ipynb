### classes

```python
class Restaurant():
    def __init__(self, restaurant_name, cusine_type):
        self.restaurant_name = restaurant_name
        self.cusine_type = cusine_type

    def describe_method(self):
        print(f"name of Restaurant is {self.restaurant_name} which is special in {self.cusine_type}")
    
    def open_restaurant(self):
        print(f"{self.restaurant_name} is open now")
        
```

```python
restaurant1 = Restaurant("Lal Qila", "Pakistani food")

restaurant1.open_restaurant()
```

```python
class User():
    def __init__(self, fName, lName):
        self.first_name = fName
        self.last_name = lName
        self.full_name = f"{lName} {fName}"
        
    def describe(self):
        print(f"The User's first name is {self.first_name} and his last name is {self.last_name}")

    def greet(self):
        print(f"Hello! {self.full_name}")
```

```python
user1 = User("Ahmed", "Rizwan")

user1.describe()
user1.greet()
```
