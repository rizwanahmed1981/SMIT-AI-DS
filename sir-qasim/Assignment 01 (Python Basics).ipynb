12) A school has following rules for grading system:

a. Below 25 - F

b. 25 to 45 - E

c. 45 to 50 - D

d. 50 to 60 - C

e. 60 to 80 - B

f. Above 80 - A

Ask user to enter marks and print the corresponding grade.

```python
#Type your code here
marks = float(input("enter your marks to check grade: "))
grade = 'not assigned'

if 80 <= marks <= 100:
    grade = "A"
if 60 <= marks < 80:
    grade = "B"
if 50 <= marks < 60:
    grade = "C"
if 45 <= marks < 50:
    grade = "D"
if 25 <= marks < 45:
    grade = "E"
if 0 <= marks < 25:
    grade = "F"
else:
    print("Invalid marks")

print("your grade is", grade)
```
